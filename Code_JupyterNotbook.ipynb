### - Check the row duplication 

```python

print('Use the keep parameter to consider only the first instance of a duplicate row to be unique: \n')
bool_series = traffic_df.duplicated(keep='first')
print('Boolean series:')
print(set(bool_series))
print('\n')


traffic_df = traffic_df[~bool_series]
```

### - Remove columns that more 90% of their rows is nulls or it's a raw data

```python

print('Check the nulls values in each columns and delete the columns with more thab 90% of it is null: \n')
for column in traffic_df.columns:
    prec = traffic_df[column].isna().sum()/len(traffic_df) * 100

    if prec > 90 or column in ['driver_age_raw','violation_raw']:
        traffic_df.drop(column, axis=1, inplace=True)
```

```python
print('check the total null values: \n', traffic_df.isna().sum())
```

### - Fill the null based on the statistical measured   

```python

print('If the column type is categorical and contains only 1 missing, \n fill it with the mode ')
for column in traffic_df.columns:
    
    if traffic_df[column].isna().sum() == 1 and traffic_df.dtypes[column] == np.object:
        traffic_df[column] = traffic_df[column].fillna(traffic_df[column].mode()[0])
  
```

```python
print('check the total null values: \n', traffic_df.isna().sum())
```

```python
# Let check pre column how many is Arrested vs. Not Arrested

print('Check how many Arrested and Not Arrested is missing in each column: \n')

nan_columns = ['driver_gender', 'driver_race' , 'violation' , 'stop_outcome', 'is_arrested', 'stop_duration']
                    
for column  in nan_columns:  
    list_index = traffic_df[traffic_df[column].isnull()].index.tolist()
    num_of_items = len(traffic_df.loc[list_index].groupby('is_arrested'))
    print(column, " null items pre is_arrested type ", str(num_of_items))
    

```

```python


print('Based on the previous information we are save to delete the empty rows: \n')
traffic_df.dropna(how='any', inplace=True)

# visualize target
visualize_target(traffic_df)
```

### - Encode the categorical data


To use SVM or logistic regression, we need to have numeric values. Encode the categorical columns.

```python
# split the dates to year, months and day


traffic_df['stop_year'] = traffic_df['stop_date'].apply(lambda x: int(x.split('/')[2]))
traffic_df['stop_month'] = traffic_df['stop_date'].apply(lambda x: int(x.split('/')[0]))
traffic_df['stop_day'] = traffic_df['stop_date'].apply(lambda x: int(x.split('/')[1]))

# drop the stop_date coluumn ( no longer needed)
traffic_df.drop('stop_date',axis=1,inplace=True)
```

```python
# convert the stop_time column from Nomianl to Interval

traffic_df['stop_time'] = traffic_df['stop_time'].apply(lambda x: int(x.split(':')[0]) + int(x.split(':')[1])/60)
```

```python
# Encode the gender as Dichotomous (0, 1)
traffic_df['driver_gender'] = traffic_df['driver_gender'].map({'M':1,'F':0})
```

```python
# Encode the search_conducted, drugs_related_stop, and is_arrested columns as Dichotomous (0, 1)
cloumns_Encoded = ['search_conducted', 'drugs_related_stop', 'is_arrested']

for cloumn in cloumns_Encoded:
    traffic_df[cloumn] =  traffic_df[cloumn].apply(lambda x: 1 if x == True else 0)
```

```python
traffic_df.head()
```

```python
print(' Encode Nominal column ')
cloumns_Encoded = ['driver_race', 'violation', 'stop_duration', 'stop_outcome']

for cloumn in cloumns_Encoded:
    temp = pd.get_dummies(traffic_df[cloumn],drop_first=False)
    if 'Other' in temp.columns:
        
        # avoid having multiple 'Other' column 
        temp['other_'+cloumn] = temp['Other']
    traffic_df = pd.concat([traffic_df,temp],axis=1)
    traffic_df.drop(cloumn,axis=1,inplace=True)


traffic_df.head()
```

### -  SVM function

```python
def SVM_hyperparameterOptimization( X_train_std, y_train,X_test, y_test, text='over Sampling' ):
    
    """

    X_train_std: X train dataset Standardized
    y_train: y train dataset (target)
    X_test: X test dataset 
    y_test: y test dataset (target)
    text: imbalanced data technique  
    
    """
    
    # kernal: Specifies the kernel type to be used in the algorithm. 
    kernels = ['linear',  'sigmoid']
    
    #C: Regularization parameter. The strength of the regularization is inversely proportional 
    #to C. Must be strictly positive. The penalty is a squared l2 penalty.
    C = [1.0, 1.5]
    
    
    # save the acc result 
    acc = []
    
    
    # for each kernal, loop over the hyperparameter and fit the model 
    for kernel in kernels:
        
        acc_kernel  = []
        
        for i in range(len(C)):

            # init the model with C and epsilon
            svm_model= svm.SVC(kernel=kernel, C=C[i], probability=True)

            # fit SVM
            svm_model.fit(X_train_std, y_train)
        

            # predict 
            predict_svm = svm_model.predict(X_test)

           # test the model accuracy 
            mae_svm = mean_absolute_error(y_test, predict_svm)
            accuracy = accuracy_score(y_test, predict_svm)

          
            # print: 
            print(' SVM with kernel {0} and {1} technique, C= {2:0.3f}  MSE: {3:0.3f}'.format(kernel, text, C[i],  mae_svm))
            print(' SVM with kernel {0} and {1} technique, C= {2:0.3f}  Acc: {3:0.3f}'.format(kernel, text, C[i],  accuracy))
            
            
            precision = precision_score(y_test, predict_svm)
            print('SVM with kernel {0} and {1} technique, C= {2:0.3f} PPTV: {3:0.3f}'.format(kernel, text, C[i], precision))
            
            TP, FP, TN, FN = confusion_matrix(y_test, predict_svm).ravel()
            print('TP: {0:0.3f} FP: {1:0.3f} TN: {2:0.3f}  FN: {3:0.3f}'.format(TP, FP, TN, FN))
            
            if (TN+FN) == 0:
                FN= 1
            NPV = TN/(TN+FN)
            print('SVM with kernel {0} and {1} technique, C= {2:0.3f} NPV: {3:0.3f}'.format(kernel, text, C[i], NPV))
            
            recall = TP/(TP+FN)
            print('SVM with kernel {0} and {1} technique, C= {2:0.3f} Recall: {3:0.3f}'.format(kernel, text, C[i], recall))
            
            specificity = TN / (TN+FP)
            print('SVM with kernel {0} and {1} technique, C= {2:0.3f} Specificity: {3:0.3f}'.format(kernel, text, C[i], specificity))
            
          
            F1 = 2* ((precision*recall)/(precision+recall))
            print('SVM with kernel {0} and {1} technique, C= {2:0.3f} F1: {3:0.3f}'.format(kernel, text, C[i], F1))


            
            print('-----------------------------------------------------------------------------------------------------------------------------------')
            
            acc_kernel.append(mae_svm)
            
        # append the acc for visualize
        acc.append(acc_kernel)

    df = pd.DataFrame(acc,
                     index=kernels,
                     columns=pd.Index([ 'MSE with C=1.0', 'MSE with C=1.5'], 
                     name='Kernels'))

    print(df)
    df.plot(kind='bar',figsize=(20,8))

    ax = plt.gca()
    pos = []

    for bar in ax.patches:
        pos.append(bar.get_x()+bar.get_width()/2.)
        ax.annotate(str(bar.get_height()), (bar.get_x() * 1.005, bar.get_height() * 1.005))



    ax.set_xticks(pos,minor=True)
    lab = []
    for i in range(len(pos)):
        l = df.values[i//len(df.index.values)][0]
        lab.append(l)



    plt.title("SVM Hyperparameter Optimization")
    plt.show()
    
```

### - SVM with OverSampled data

### 2. Logistic Regression function

```python
def LR_hyperparameterOptimization( X_train_std, y_train, X_test, y_test, text='over Sampling' ):
    
    """

    X_train_std: X train dataset Standardized
    y_train: y train dataset (target)
    X_test: X test dataset 
    y_test: y test dataset (target)
    text: imbalanced data technique  
    
    """
    
    # solver: Specifies the kernel type to be used in the algorithm. 
    solvers = ['newton-cg', 'lbfgs', 'liblinear']
    
  
    acc = []
   

    # for each solver, loop over the hyperparameter and fit the model 
    for solver in solvers:
        
   
        # init the model with diffrernt solver
        LogisticRegression_model = LogisticRegression(solver=solver, random_state=42)

        # fit SVM
        LogisticRegression_model.fit(X_train_std, y_train)

        # predict 
        predict_lr = LogisticRegression_model.predict(X_test)

         # test the model accuracy 
        mae_LAR = mean_absolute_error(y_test, predict_lr)
        accuracy = accuracy_score(y_test, predict_lr)
    
        # print
        print(' Logistic Regression with {0} and solver {1}, MSE: {2:0.3f}'.format(text, solver, mae_LAR))
        print(' Logistic Regression with {0} and solver {1}, accuracy: {2:0.3f}'.format(text, solver, accuracy))

        
        precision = precision_score(y_test, predict_lr)
        print('SVM with kernel {0} and {1} technique, C= {2:0.3f} PPTV: {2:0.3f}'.format(text, solver, precision))

        TP, FP, TN, FN = confusion_matrix(y_test, predict_lr).ravel()
        print('TP: {0:0.3f} FP: {1:0.3f} TN: {2:0.3f}  FN: {3:0.3f}'.format(TP, FP, TN, FN))

        if (TN+FN) == 0:
            FN= 1
        NPV = TN/(TN+FN)
        print('Logistic Regression with {0} and solver {1} NPV: {2:0.3f}'.format(text, solver, NPV))

        recall = TP/(TP+FN)
        print('Logistic Regression with {0} and solver {1} Recall: {2:0.3f}'.format(text, solver, recall))

        specificity = TN / (TN+FP)
        print('Logistic Regression with {0} and solver {1}Specificity: {2:0.3f}'.format(text, solver, specificity))


        F1 = 2* ((precision*recall)/(precision+recall))
        print('Logistic Regression with {0} and solver {1} F1: {2:0.3f}'.format(text, solver, F1))
    
        print('-----------------------------------------------------------------------------------------------------------------------------------')


            
        # append the acc for visualize
        acc.append(mae_LAR)
                                  

    df = pd.DataFrame(acc,
                     index=solvers,
                     columns=pd.Index([ 'acc']))


    df.plot(kind='bar',figsize=(20,8))

    ax = plt.gca()
    pos = []

    for bar in ax.patches:
        pos.append(bar.get_x()+bar.get_width()/2.)
        ax.annotate(str(bar.get_height()), (bar.get_x() * 1.005, bar.get_height() * 1.005))



    ax.set_xticks(pos,minor=True)
    lab = []
    for i in range(len(pos)):
        l = df.values[i//len(df.index.values)][0]
        lab.append(l)



    plt.title("SVM Hyperparameter Optimization")
    plt.show()
    
```
